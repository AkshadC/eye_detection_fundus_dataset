```python
import os
import random
import shutil

# Define paths
fundus_folder = "Dataset/fundus/test"
fundus_split_folder = "Fundus/split"
train_folder = os.path.join(fundus_split_folder, "val")
test_folder = os.path.join(fundus_split_folder, "test")

# Create split folder if it doesn't exist
os.makedirs(fundus_split_folder, exist_ok=True)

# Create train and test folders inside split folder
os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# Get list of bird class folders
fundus_class_folders = [folder for folder in os.listdir(fundus_folder) if os.path.isdir(os.path.join(fundus_folder, folder))]

# Create class folders in train and test folders
for class_folder in fundus_class_folders:
    train_class_folder = os.path.join(train_folder, class_folder)
    test_class_folder = os.path.join(test_folder, class_folder)
    os.makedirs(train_class_folder, exist_ok=True)
    os.makedirs(test_class_folder, exist_ok=True)


# Split dataset into train and test
for fundus_class_folder in fundus_class_folders:
    class_images = os.listdir(os.path.join(fundus_folder, fundus_class_folder))
    random.shuffle(class_images)  # Shuffle images
    split_index = int(0.67 * len(class_images))  

    # Randomly split images into train and test
    train_images = class_images[:split_index]
    test_images = class_images[split_index:]

    # Move images to train folder
    for image in train_images:
        src = os.path.join(fundus_folder, fundus_class_folder, image)
        dst = os.path.join(train_folder, fundus_class_folder, image)
        if not os.path.exists(dst):
            shutil.copy(src, dst)

    # Move images to test folder
    for image in test_images:
        src = os.path.join(fundus_folder, fundus_class_folder, image)
        dst = os.path.join(test_folder, fundus_class_folder, image)
        if not os.path.exists(dst):
            shutil.copy(src, dst)

print("Dataset split into train and test folders successfully.")
```
