```python
import math
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from matplotlib.animation import FuncAnimation
data_points = [[7,7],
               [5,9],
               [6,9],
               [7,9],
               [5,11],
               [5,3],
               [6,1],
               [6,2],
               [7,1],
               [7,2],
               [7,3],
               [8,4],
               [8,6],
               [9,3],
               [9,4],
               [9,5],
               [10,4],
               [10,5],
               [10,6],
               [9,7]]
```

```python
def plot_kmeans(points, n_clusters=2, max_iter=10):
    fig, ax = plt.subplots()
    colors = ['r', 'g', 'b', 'y', 'c', 'm']

    def update(frame):
        ax.clear()
        ax.set_title(f'K-Means Clustering with {n_clusters} Clusters\nIteration {frame + 1}')
        kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=frame + 1, n_init=1, random_state=0)
        kmeans.fit(points)
        labels = kmeans.labels_
        for i in range(n_clusters):
            cluster_points = points[labels == i]
            ax.scatter(cluster_points[:, 0], cluster_points[:, 1], s=50, c=colors[i % len(colors)], label=f'Cluster {i + 1}')
        ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='black', marker='X', label='Centroids')
        ax.legend()

    def frame_generator():
        for i in range(max_iter):
            yield i

    ani = FuncAnimation(fig, update, frames=frame_generator, repeat=False, cache_frame_data=False)
    plt.show()

# Plot K-Means for 2 clusters and show the animation
plot_kmeans(data_points, n_clusters=2)
```

```python
def get_distance(c, j):
    return math.sqrt(math.pow(c[0] - j[0], 2) + math.pow(c[1] - j[1], 2))
```

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt


# Load the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()

# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0

# Reshape the data to include the channel dimension
train_images = train_images.reshape((train_images.shape[0], 28, 28, 1))
test_images = test_images.reshape((test_images.shape[0], 28, 28, 1))
model = models.Sequential()
model.add(layers.Flatten())
model.add(layers.Dense(784, activation='sigmoid'))
model.add(layers.Dense(10, activation='softmax'))

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(train_images, train_labels, epochs=5, 
                    validation_data=(test_images, test_labels))

test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f"Test accuracy: {test_acc}")


plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')

plt.show()
```
